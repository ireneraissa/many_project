```python
from pypedriver import Client
pipedrive = Client('8fd567f2a2963d2f01c0572d1ea8b29433fbec7d')
```

```python
Org=[]
Orgs=[]
Orgid=[]
organizations=pipedrive.Organization().fetch_all(filter_id=None, start=0)
for organization in organizations:
    Org.append(organization.name)
    Orgs.append(organization.secteur_normalisé)
    Orgid.append(organization.id)
```

```python
from unidecode import unidecode
Org_name=[ ]
for i in range(len(Org)):
    
    Org_name.append(unidecode(Org[i]).lower())
```

```python
sect_organisation_id=[]
for i in range(len(Org)):
    alt=  {'secteur':Orgs[i],'organis':Org_name[i],'id':Orgid[i]}
    sect_organisation_id.append(alt)

sect_organisation_id
```

```python
fichier_csv=pd.read_csv('data.csv')
```
