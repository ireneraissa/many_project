```python
#I do the same things like in the previous cell , here is for all the dataframe in a list GRoupee but i delete 
#line with the same signal(sig) and same day
index_repeat_line=[]

for i in range(len(firefly_signal_order_device_id)) :
    
    
    index_line_to_cancel=[] 
 
    for j in range((len(firefly_signal_order_device_id[i]))-1):
        signal=firefly_signal_order_device_id[i]['sig']
        date=firefly_signal_order_device_id[i]['day']
        id_device=firefly_signal_order_device_id[i]['device_id']
        
        if (signal[signal.index[j]]==firefly_signal_order_device_id[i]['sig'][signal.index[j+1]]
             and [date.index[j]]==date[date.index[j+1]]):
            
            #print(1)
             index_line_to_cancel.append(id_device.index[j+1])
      
            
            
            
    index_repeat_line.append(index_line_to_cancel)
    
```

```python

Hour=[]
for i in range(len(firefly_signal_order_device_id)) :
    
    
    
    hour=[] 
    for j in range((len(firefly_signal_order_device_id[i]))-1):
        signal=firefly_signal_order_device_id[i]['sig']
        date=firefly_signal_order_device_id[i]['day']
        id_device=firefly_signal_order_device_id[i]['device_id']
        
        if (date[date.index[j]]!=date[date.index[j+1]] ):
            
            hour.append(id_device.index[j])
            
    if date[date.index[len(firefly_signal_order_device_id[i])-1]] not in hour:
        hour.append(id_device.index[len(firefly_signal_order_device_id[i])-1])
      
            
            
    Hour.append(hour)
```
