```python
import requests
from bs4 import BeautifulSoup
url='http://pagesjaunesdusenegal.com'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html,'html5lib')
```

```python
table=soup.text
```

```python
table
```

```python
url_veritable=table.replace('\n\twindow.location.href = "','').replace('";\n','')
```

```python
answer = requests.get(url_veritable)
lxml = answer.content
soupe = BeautifulSoup(lxml,'html5lib')
```

```python
table_veritable=soupe.find('div',attrs={'class':'row background-menu'})
```

```python
table_veritable.findAll('li')
```

```python
lien=[]
for row in table_veritable.findAll('div'):
    for cell in row.findAll('a'):
        lien.append(cell)
        
```

```python
lien
```

```python
lien_recuperer=[]
for line in lien:
    if line['href']== ' ':
        continue
    else: 
        lien_recuperer.append(line['href'])
```

```python
lien_recuperer
```

```python
import numpy  as np
```

```python
lien_util=lien_recuperer[0:np.int(len(lien_recuperer)/2)]
```

```python
len(lien_util)
```

```python
lien_util_bank=lien_util[20]
```

```python
lien_util_bank
```

```python
html_bank=BeautifulSoup(requests.get(lien_util_bank).content,'html5lib')
```

```python
html_bank
```

```python
research=html_bank.find('div',attrs={'class': 'pagination'}) 
```

```python
research
```

```python
research_find=research.findAll('a')
```

```python
research_find
```

```python
lien_different_bank=[]
for line in research_find:
    if line['href']== ' ':
        continue
    else: 
        lien_different_bank.append(line['href'])
```

```python
lien_different_bank
```

```python
lien_different_banks=[]
for line in lien_different_bank :
    lien_different_banks.append(line.replace(line,'http://pagesjaunesdusenegal.com'+line))  
lien_different_banks
```
