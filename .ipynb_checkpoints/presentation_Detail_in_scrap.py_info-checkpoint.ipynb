```python
import requests
from bs4 import BeautifulSoup
url = 'http://www.showmeboone.com/sheriff/JailResidents/JailResidents.asp'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html,'lxml')
```

```python
table = soup.find('tbody', attrs={'class': 'stripe'})
lien_detail =[]
for line in table.findAll("a",):
    if line['href']== ' ':
        continue
    else: 
        lien_detail.append(line['href'])
```

```python
lien_detail_prisonnier=lien_detail[0:50]
url_Details=[]
for line in lien_detail_prisonnier :
    url_Details.append(line.replace(line,'https://report.boonecountymo.org/mrcjava/servlet/'+line))  
```

```python
url_Details
```

```python
get_all_html_Detail=[]
for i in range(len(url_Details)):
    requests.get(url_Details[i]).content
    #for line in re:
      #  rei.append(re.content)
#for line in re:
    get_all_html_Detail.append(BeautifulSoup(requests.get(url_Details[i]).content,'html5lib'))
```

```python
get_all_html_Detail
```

```python
lists_row_all=[]
for i in range(len(url_Details)):
    
    table_all=get_all_html_Detail[i].find('div',attrs={'id' : "detaineeInfo"})
    table_all_left= table_all.find('tbody',attrs={'class' : 'single'})
    lists_row=[]
    for row in table_all_left.findAll('tr'):        
        #list_cells_all = [] i remove this because i dont want my information in list ,
        #but now i will loose what is inside 'td'
        for cell in row.findAll('td'):
            #list_cells_all.append(cell.text)
            list_cells_all=cell.text.replace('W', 'white')
        lists_row.append(list_cells_all)
    lists_row_all.append(lists_row)
```

```python
#lists_row_all[0]
for line in lists_row_all:
    for i in range(len(line)):
        if line[i]=='B':
            line[i]='Black'
```

```python
lists_row_all[2]
```

```python
lists_row_all_name=[]
for i in range(len(url_Details)):
    table_all_name=get_all_html_Detail[i].find('div',attrs={'id' : "inmateName"})
    lists_row_all_name.append(table_all_name.text)
lists_row_all_name
```
