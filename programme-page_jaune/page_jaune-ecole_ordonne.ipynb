```python
while browser.find_element_by_class_name("pagination").find_elements_by_tag_name('a')[len(browser.find_element_by_class_name("pagination").find_elements_by_tag_name('a'))-1].text == '>' : #in browsers :
    
    driver= browser.find_element_by_class_name("pagination")
    
    
    
    for line in driver.find_elements_by_tag_name('a'):
        
        while line.text =='>':
            
            line.send_keys(Keys.ENTER)
            

             
            browsers = browser.find_element_by_xpath('/html/body/div/div[4]')
            for line in browsers.find_elements_by_tag_name('button'):
                line.send_keys(Keys.ENTER)
             
            for div in browsers.find_elements_by_class_name('desk-result'):
                all_info=[]
                all_info.append(div.text)
                als_info.append(all_info)
                
    als_info.remove(['Accueil > Résultats de recherche'])
```

```python
all_infos=[]
for i in range(len(als_info)):
    text=als_info[i][0].replace('\n1\n2\n3\n4\n5\n', '')
    text1=text.replace('\n1\n2\n3\n4\n5', '  ')
    all_infos.append(text1)
```

```python
tests=[]
for i in range(len(all_infos)):
    tests.append(all_infos[i].split('Notez ce professionnel'))
```

```python
decompose=[]
for i in range(len(tests)):
    for j in range(len(tests[i])):
        decompose.append(tests[i][j])
```

```python
testss=[]
for i in range(len(decompose)):
    testss.append(decompose[i].split('\n'))
    
decomposition=[]
for i in range(len(testss)):
    tes=[]
    for j in range(len(testss[i])):
        
        tes.append(testss[i][j])
    decomposition.append(tes)
```

```python
decomposition
```

```python
for line in decomposition:
    if line ==['  '] :
        decomposition.remove(line)
```

```python
decomposition
```

```python
for line in decomposition:
    for row in line:
        if row =='' :
            line.remove(row)
        if row =='Voir fiche | Présentation' :
            line.remove(row)
```

```python
for line in decomposition:
    for row in line:
        if row =='Voir fiche | Visiter le site' :
            line.remove(row)
        if row =='   Afficher le numéro' :
            line.remove(row)
        if row =='Logo' :
            line.remove(row)
```
